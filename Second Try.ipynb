```python
import re
def clean_text(text):
    text = text.strip()
    text = text.replace('\xa0', ' ')
    text = text.replace('\n', ' ').replace('\t', ' ')
    text = re.sub(' +', ' ', text)
    return text
```

```python
raw_data['Cleaned_Resume'] = raw_data['Resume_str'].apply(clean_text)
```

```python
from bs4 import BeautifulSoup

def extract_name(html_content, code):
    soup = BeautifulSoup(html_content, 'html.parser')
    
    name_span = soup.find('span', {'id': lambda x: x and x.endswith(code)})
    if name_span:
        return name_span.get_text(strip=True)
    else:
        return None
```

```python
html_code = ['LNAM1', 'DGRE1', 'STUY1']
for code in html_code:
    HR_data[code] = HR_data['Resume_html'].apply(lambda x: extract_name(x, code))
```

```python
def standardize_degree(degree):
    degree = degree.lower().replace('.', '').replace(' of ', ' ').replace('science', '').strip()
    
    bachelor_variations = ['bachelor', 'bs', 'bsc', 'ba', 'b a']
    master_variations = ['master', 'ms', 'msc', 'ma', 'm a', 'mba', 'm b a']
    doctorate_variations = ['doctor', 'phd', 'ph d', 'doctorate']
    
    if any(bach in degree for bach in bachelor_variations):
        return 'Bachelor'
    elif any(mast in degree for mast in master_variations):
        return 'Master'
    elif any(doc in degree for doc in doctorate_variations):
        return 'Doctorate'
    else:
        return 'Other'
```

```python
HR_data['DGRE1'] = HR_data['DGRE1'].apply(standardize_degree)

HR_data = HR_data[HR_data['DGRE1'].isin(['Bachelor', 'Master', 'Doctorate'])]
```

```python
def extract_exp(html_content, code):
    soup = BeautifulSoup(html_content, 'html.parser')
    
    name_span = soup.find('span', {'id': lambda x: x and x.endswith(code)})
    if name_span:
        if(name_span.get_text(strip=True) == 'Current'):
            return '01/2021'
        else:
            return name_span.get_text(strip=True)
    else:
        return None
```

```python
code_exp = exp_data.drop(columns = ('ID')).columns.tolist()

for code in code_exp:
    exp_data[code] = HR_data['Resume_html'].apply(lambda x: extract_exp(x, code))
```

```python
def preprocess_date(date_str):
    if pd.isna(date_str):
        return 'Unknown'
    date_formats = [
        '%m/%Y',           
        '%B %Y',           
        '%b %Y'            
    ]
    
    for fmt in date_formats:
        try:
            parsed_date = pd.to_datetime(date_str, format=fmt)
            return parsed_date.strftime('%m-%Y')  
        except ValueError:
            pass  
    
    return 'Unknown'
```

```python
for col in exp_data.columns:
    if col != 'ID':
        exp_data[col] = exp_data[col].apply(preprocess_date)

for col in exp_data.columns:
    if col != 'ID':
        exp_data[col] = pd.to_datetime(exp_data[col], errors='coerce')
```

```python
date_columns = [col for col in exp_data.columns if 'JSTD' in col or 'EDDT' in col]
exp_data[date_columns] = exp_data[date_columns].apply(pd.to_datetime)

exp_data['Total_Duration_Years'] = np.nan

for index, row in exp_data.iterrows():
    total_duration_seconds = 0
    for i in range(1, 11):
        start_date_col = f"JSTD{i}"
        end_date_col = f"EDDT{i}"
        start_date = row[start_date_col]
        end_date = row[end_date_col]

        if not pd.isnull(start_date) and not pd.isnull(end_date):
            duration = end_date - start_date
            total_duration_seconds += duration.total_seconds()

    total_duration_years = total_duration_seconds / (365.25 * 24 * 3600)
    exp_data.at[index, 'Total_Duration_Years'] = total_duration_years
```

```python
exp_data['Total_Duration_Years'] = round(exp_data['Total_Duration_Years'], 1)
```

```python
def extract_primary_position(title):
    positions = re.split(r'[,&/]', title)    
    primary_position = positions[0].strip()
        
    return primary_position
```
